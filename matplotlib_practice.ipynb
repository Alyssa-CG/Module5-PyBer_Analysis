```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot with MATLAB
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() OBJECT-ORIENTED APPROACH
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
# NOTE: To change figure-level attributes 
# (such as axis labels, a title, or a legend) 
# or save the figure as an image, use fig = plt.figure().


fig = plt.figure()
ax = fig.add_subplot()

#above returned blank plot area

ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() Quick way
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Add the legend.
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# NOTE, the"o" makes it a SCATTERPLOT instead of a LINE GRAPH
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(x_axis, y_axis)
```

```python
# SKILL DRILL
# Using the Matplotlib MATLAB plotting approach:

# Change the color of the markers to red.
# Add a legend for the city of Chicago.
# Add a title and axes labels.
# Switch the axis so that the Fare($) data is on the x-axis.
plt.scatter(y_axis,x_axis, color="red",label="Chicago")
plt.legend(loc="best")

plt.title("PyBer Fare by Month")
plt.xlabel("Fare ($)")
plt.ylabel("Date")

# Add a limit for the x-axis data.
plt.xlim(0,50)

# Invert the y-axis so that January is at the top
plt.gca().invert_yaxis()
```

```python
# Change size in scatterplot. Size must be equal to 
# a feature of the data, example Fare (y-axis).

plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Iterated through each data point, for size (s) multiplied by 3 and plotted.

y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Do the same thing with LIST COMPREHENSION

plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# OBJECT ORIENTED INTERFACE

fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# SKILL DRILL Using the object-oriented approach:

# Change the color of the markers to sky blue.
# Change the size of the markers to 5 times each data point.
# Make the color 20% transparent.
# Add a black edge color to the circles.
# Make the linewidth of the circles 2 points.
# Add a legend for the city of Boston.

fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i*5 for i in y_axis],label="Boston",
           color="skyblue",alpha=0.8,
           edgecolor="black",linewidth=2)

plt.legend(loc="best")

# Add a title.
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")

# Switch the axis so that the Fare($) data is on the x-axis. (switched above).
# Add a limit for the x-axis data.
ax.set_xlim(0,50)
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Pie Chart using OBJECT-ORIENTED INTERFACE

fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)
plt.show()
```

```python
# SKILL DRILL
# Using the object-oriented approach:

# Add a percentage to one decimal place to each wedge of the pie. --
# Increase the figure size to 8x8.--
# Explode the two highest percentage months.--
# Add a shadow. --
# Add a start angle so that January is at the top.--
# Reverse the order so that the month order is in a clockwise direction. --
# Add new colors of your own choosing or use the colors --
# from the previous pie chart. --

fig, ax = plt.subplots()

explode_values=(0,0,0,0,0.2,0,0.2,0,0,0,0,0)
colors= ("olivedrab","yellowgreen","darkolivegreen","lawngreen",
        "darkseagreen","forestgreen","limegreen","green",
        "lime","seagreen","springgreen","greenyellow")

fig.set_size_inches(8,8)
ax.pie(y_axis,labels=x_axis,
    explode=explode_values,
    colors=colors,
    autopct='%.1f%%',
    shadow=True,
    counterclock=False,
    startangle=90)

plt.show()
```
